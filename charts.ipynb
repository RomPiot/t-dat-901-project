```python
data = open_dataset('KaDo_clean.csv')
data
```

```python
segmBuyer = customersData['segm_buyer_type'].value_counts()
segmBuyerLabels = list(segmBuyer.keys().values)
segmBuyerValues = segmBuyer.to_list()
segmBuyerValues.insert(0, segmBuyerValues.pop())
segmBuyerLabels.insert(0, segmBuyerLabels.pop())

segmBuyer
```

```python

plt.style.use(['seaborn-talk'])

ypos = np.arange(len(segmBuyerLabels))
plt.xticks(ypos, segmBuyerLabels)
plt.title("RÃ©partition par type d'acheteur")
plt.bar(ypos, segmBuyerValues)
plt.legend(["Nombre de client"])
```
